
 Imputing a Python script which is ATS-friendly

```python
from fpdf import FPDF

class PDF(FPDF):
    def header(self):
        self.set_font('Arial', 'B', 12)
        self.cell(0, 10, 'L.O - Data Analyst', 0, 1, 'C')
        self.ln(10)

    def section_title(self, title):
        self.set_font('Arial', 'B', 12)
        self.cell(0, 10, title, 0, 1)
        self.ln(5)

    def section_content(self, content):
        self.set_font('Arial', '', 10)
        self.multi_cell(0, 10, content)
        self.ln()

pdf = PDF()
pdf.add_page()

pdf.set_font('Arial', '', 10)

# Personal Information
pdf.section_title('Contact Information:')
pdf.section_content('Email: ol@gmail.com\nPhone: +234 907 000 XXXX\nLinkedIn: L O\nGitHub: thecoderblue7')

# Education
pdf.section_title('Education:')
pdf.section_content('BSc in Biochemistry, Crawford University, Ogun State - 2018')

# Work Experience
pdf.section_title('Work Experience:')
pdf.section_content('- Data Science/Analytics Apprentice at FLiT Inc. (October 2023 - Present)\n'
                    '- Data Analyst Associate Intern at Excelerate Internships (December 2023 - Present)\n'
                    '- Research Analyst Freelance (January 2023 - Present)')

# Certifications and Programs
pdf.section_title('Certifications and Programs:')
pdf.section_content('- Google Analytics Specialization Certificate, Coursera - October 2023\n'
                    '- FGN-ALAT Digital Skillnovation Program for Data Science/Analytics')

# Skills
pdf.section_title('Skills:')
pdf.section_content('Spreadsheets, R, GitHub, Python, SQL, Tableau, Pivot Table')

# Languages
pdf.section_title('Languages:')
pdf.section_content('English, French')

# References
pdf.section_title('References:')
pdf.section_content('Available upon request.')

pdf.output('L_O_Resume.pdf')
```
